```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.typingstudy.com/typingtest/1/The_Fox_and_the_Grapes'
page = requests.get(url)
html = page.text
soup = BeautifulSoup(html, "html")
print(soup.prettify())
```

# Query Paragraph Title List

```python
ls=[]
min_character = 750

for element in soup.select("#stories_select option"):
    title = element.get_text().split()
    if int(title[-1][1:-1]) > min_character: # filter only paragraph with long text
        ls.append([element.get("value"), title[:-1]])
ls
```

# Query Paragraph Text for each Item in List

```python
data = []
for item in ls:
    value = item[0]
    title = "_".join(item[1])
    url = f'https://www.typingstudy.com/typingtest/{value}/{title}'
    page = requests.get(url)
    html = page.text
    soup = BeautifulSoup(html, "html")
    text = soup.select("#type_text")[0].get("value")[:-1]
    data.append([url, text])

df = pd.DataFrame(
    data=data,
    columns=['URL','Paragraph']
)
df
```

```python
df.to_csv("dataset/paragraph.csv")
```
